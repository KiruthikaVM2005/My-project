```python
import pandas as pd
```

```python
df=pd.read_csv('Crop_recommendation.csv')
df.head()
```

```python
df.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
x=df.drop('label',axis=1)
y=df['label']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=RandomForestClassifier(n_estimators=100)
model.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
accuracy=accuracy_score(y_test,predictions)
print(accuracy)
```

```python
crop_prices={'rice':20, 'maize':18,'banana':30, 'mango':50, 'cotton':60,'coconut':35}
def estimate_yield(crop):
  yield_data={'rice':3000, 'maize':2500,'banana':20000, 'mango':10000, 'cotton':2000,'coconut':10000}
  return yield_data.get(crop,2000)
```

```python
def calculate_profit(crop):
  price=crop_prices.get(crop,20)
  yield_amt=estimate_yield(crop)
  profit=price*yield_amt
  return profit
```

```python
region_data = {
    'Tamil Nadu': {'temperature': 30, 'humidity': 70, 'rainfall': 150},
    'Kerala': {'temperature': 28, 'humidity': 85, 'rainfall': 300},
    'Punjab': {'temperature': 25, 'humidity': 60, 'rainfall': 100}
}
```

```python
region = input("Enter Region: ")
region_values = region_data.get(region, {'temperature': 25,'humidity': 60,'rainfall': 100})
N = float(input("Enter Nitrogen: "))
P = float(input("Enter Phosphorus: "))
K = float(input("Enter Potassium: "))
ph = float(input("Enter pH: "))
temperature = input("Enter Temperature (press Enter to use region value): ")
humidity = input("Enter Humidity (press Enter to use region value): ")
rainfall = input("Enter Rainfall (press Enter to use region value): ")
temperature = float(temperature) if temperature else region_values['temperature']
humidity = float(humidity) if humidity else region_values['humidity']
rainfall = float(rainfall) if rainfall else region_values['rainfall']
user_data = pd.DataFrame([[N, P, K, temperature, humidity, ph, rainfall]],
                         columns=['N','P','K','temperature','humidity','ph','rainfall'])
probs = model.predict_proba(user_data)[0]

top3_idx = probs.argsort()[-3:][::-1]
top3_crops = model.classes_[top3_idx]

print("\nTop 3 Crops:\n")

for crop in top3_crops:
    profit = calculate_profit(crop)
    print("Crop:", crop)
    print("Estimated Profit:", profit)
    print("------")
```
